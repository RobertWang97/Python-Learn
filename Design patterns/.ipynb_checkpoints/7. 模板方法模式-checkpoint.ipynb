<h1>模板模式</h1>
在模板模式（Template Pattern）中，一个抽象类公开定义了执行它的方法的方式/模板。它的子类可以按需要重写方法实现，但调用将以抽象类中定义的方式进行。这种类型的设计模式属于行为型模式。

介绍
<h4>意图</h4>：定义一个操作中的算法的骨架，而将一些步骤延迟到子类中。模板方法使得子类可以不改变一个算法的结构即可重定义该算法的某些特定步骤。

<h4>主要解决</h4>：一些方法通用，却在每一个子类都重新写了这一方法。

<h4>何时使用</h4>：有一些通用的方法。

<h4>如何解决</h4>：将这些通用算法抽象出来。

<h4>关键代码</h4>：在抽象类实现，其他步骤在子类实现。

<h4>应用实例</h4>： 1、在造房子的时候，地基、走线、水管都一样，只有在建筑的后期才有加壁橱加栅栏等差异。 2、西游记里面菩萨定好的 81 难，这就是一个顶层的逻辑骨架。 3、spring 中对 Hibernate 的支持，将一些已经定好的方法封装起来，比如开启事务、获取 Session、关闭 Session 等，程序员不重复写那些已经规范好的代码，直接丢一个实体就可以保存。

<h4>优点</h4>： 1、封装不变部分，扩展可变部分。 2、提取公共代码，便于维护。 3、行为由父类控制，子类实现。

<h4>缺点</h4>：每一个不同的实现都需要一个子类来实现，导致类的个数增加，使得系统更加庞大。

<h4>使用场景</h4>： 1、有多个子类共有的方法，且逻辑相同。 2、重要的、复杂的方法，可以考虑作为模板方法。

<h4>注意事项</h4>：为防止恶意操作，一般模板方法都加上 final 关键词。<br />



```python
from abc import ABCMeta, abstractmethod
```

```python
class Compiler(metaclass=ABCMeta):
    @abstractmethod
    def collectSource(self):
        print('aaa')
        pass
    
    @abstractmethod
    def compileToObject(self):
        pass
    
    @abstractmethod
    def run(self):
        pass
    
    def compileAndRun(self):
        self.collectSource()
        self.compileToObject()
        self.run()

class iOSCompiler(Compiler):
    def collectSource(self):
        print("iOS collect source!")
        
    def compileToObject(self):
        print("iOS compile to object!")
    
    def run(self):
        print("iOS compile and run!")
```

```python
iOS = iOSCompiler()
iOS.compileAndRun()
```
