```python
class Model():
    def __init__(self):
        self.model = {"a":1, "b":2, "c":3}
```

```python
class View():
    def __init__(self):
        pass
    
    def show(self, model):
        print(model['a'])
```

```python
class Controller():
    def __init__(self):
        self.model = Model()
        self.view = View()
        return
    
    def run(self):
        self.view.show(self.model.model)
    
```

```python
c = Controller()
c.run()
```
