```python
from abc import ABCMeta, abstractmethod
class State(metaclass=ABCMeta):
    @abstractmethod
    def Handle(self):
        pass
    
class ConcreteStateB(State):
    def Handle(self):
        print("Concrete State B")
        return
    
class ConcreteStateA(State):
    def Handle(self):
        print("Concrete State A")
        return
    
class Context(State):
    def __init__(self):
        self.state = None
        return
    
    @property
    def getState(self):
        return self.state
    
    def setState(self, state):
        self.state = state
        
    def Handle(self):
        self.state.Handle()
        
context = Context()
stateA = ConcreteStateA()
stateB = ConcreteStateB()

context.setState(stateA)
context.Handle()
    
```
